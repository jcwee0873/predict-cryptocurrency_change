```python
code_list = ['ADA', 'XTZ', 'XRP', 'XLM', 'DOGE', 'MATIC', 'AXS', 'ETC',
             'SAND', 'HBAR', 'LINK', 'VET', 'SOL', 'BTC', 'ETH', 'TRX', 'CRO',
             'THETA', 'DOT', 'MANA', 'LTC']
name_list = ['Cardano', 'Tezos', 'XRP', 'stellar', 'Dogecoin', 'Polygon', 'Axie-Infinity', 'Ethereum-Classic',
            'The-Sandbox', 'Hedera', 'Chainlink', 'Vechain', 'Solana', 'Bitcoin', 'Ethereum', 'Tron', 'Crypto-com-coin',
            'theta', 'polkadot-new', 'decentraland', 'litecoin']

def get_name(code) :
    temp_df = pd.DataFrame({'Code': code_list, 'Name':name_list})
    return temp_df[temp_df.Code == code]['Name'].values[0].lower()

def get_investing_historical(name):
    temp_dict = {
        'Date' : [],
        'Price_investing' : [],
    }
    
    url = f'https://www.investing.com/crypto/{name}/historical-data'
    header = {
      "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
      "X-Requested-With": "XMLHttpRequest"
    }
    
    res = requests.get(url, headers = headers)
    soup = BeautifulSoup(res.text, 'lxml')
    
    for tag in soup.select('table.historicalTbl > tbody > tr'):
        temp = tag.text.split('\n')
        temp_dict['Date'].append(temp[1])
        temp_dict['Price_investing'].append(temp[2])
    df = pd.DataFrame(temp_dict)
    df['Date'] = pd.to_datetime(df['Date'])
    return df.set_index('Date')
```

```python
for code in code_list :
    print(code)
    print(get_investing_historical(get_name(code)))
```
